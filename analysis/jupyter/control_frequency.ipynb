```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
data = []
names = []

meanprops = {
    "marker": "o",
    "markersize": 5,
}

for filename in os.listdir(os.path.join(r"C:\Users\anjaz\OneDrive\Desktop\JAKA\Eduexo_PC\analysis\frequency_data\test")):
    if filename.startswith("frequency_data"):
        with open(os.path.join(r"C:\Users\anjaz\OneDrive\Desktop\JAKA\Eduexo_PC\analysis\frequency_data\test", filename), "r") as file:
            a = [float(line.strip()) for line in file]
        data.append(a)
        suffix = filename[len("frequency_data_"):-len(".txt")]
        names.append(str(suffix))

plt.boxplot(data, vert=False, labels=names, manage_ticks=True, showfliers=False, showmeans=True, meanprops=meanprops)
plt.title(f'Control frequency of main program on PC and EXO')
plt.xlabel("Frequency [Hz]")
plt.ylabel("experiment")
plt.show
```
