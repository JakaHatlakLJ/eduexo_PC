```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Trapezoid

```python
time = 500 # [ms]
max_y = 1
portion = 5
x, step = np.linspace(0, time, 101, retstep=True)  # Create x with 101 elements for frequency of 200Hz
l = len(x)  
y = []
edge = math.ceil(l/portion)    
middle =  math.floor(l*(portion-1)/portion)
k = max_y / (step * edge)

for i in range(l):
    if i < edge:
        y.append(k*x[i])
    elif i < middle:
        y.append(max_y)
    else:
        y.append(-k*(x[i] - time))

plt.plot(x,y)
plt.show()
```

Triangle

```python
time = 500 # [ms]
max_y = 1
x = np.linspace(0, time, 101)
l = len(x)  
y = []
k = max_y * 2 / time


for i in range(l):
    if i < math.floor(l/2):
        y.append(k*x[i])
    else:
        y.append(-k*(x[i] - 500))

plt.plot(x,y)
plt.show()
```

Sinus

```python
time = 500 # [ms]
max_y = 1
x = np.linspace(0, time, 101)
y = max_y/2 * (-np.cos(2*np.pi*x / time) + 1)

plt.plot(x,y)
plt.show()
```

Pulse

```python
time = 500 # [ms]
max_y = 1
pulse_width = 8
x = np.linspace(0, time, 101) 
l = len(x) 
y = [] 

for i in range(l):
    if pulse_width % 2 == 0:
        if i < (pulse_width/2 - 0.5)*l/pulse_width or i > (pulse_width/2 + 0.5)*l/pulse_width:
            y.append(0)
        else:
            y.append(max_y)
    else:
        if i < math.floor(math.floor(pulse_width/2)*l/pulse_width) or i > math.floor(math.ceil(pulse_width/2)*l/pulse_width):
            y.append(0)
        else:
            y.append(max_y)

plt.plot(x,y)
plt.show()
```

Smoothed Trapezoid

```python
time = 500 # [ms]
max_y = 1
portion = 3
x, step = np.linspace(0, time, 101, retstep=True)
l = len(x)  
y = []
edge = math.ceil(l/portion)    
middle =  math.floor(l*(portion-1)/portion)
k = max_y / (step * edge)

for i in range(l):
    if i < edge:
        y.append((np.cos(np.pi*x[i]*portion / time + np.pi) + 1) * max_y/2)
    elif i < middle:
        y.append(max_y)
    else:
        if portion % 2 == 0:
            y.append((-np.cos(np.pi*x[i]*portion / time) + 1) * max_y/2)
        else:    
            y.append((np.cos(np.pi*x[i]*portion / time) + 1) * max_y/2)

plt.plot(x,y)
plt.show()
```
