```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Trapez

```python
x = np.linspace(0, 500, 51)  # Create x with 51 elements
l = len(x)  
y = [] 
portion = 5
max = 500
edge = l/portion
middle = l*(portion-1)/portion
k = 1


for i in range(l):
    if i < math.floor(edge):
        y.append(k*x[i])
    elif i < math.ceil(middle):
        y.append(k*x[math.floor(edge)])
    else:
        y.append(-k*(x[i] - 500))

print(y)

plt.plot(x,y)
plt.show()
```
